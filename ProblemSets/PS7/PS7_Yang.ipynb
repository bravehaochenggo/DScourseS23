```python
import numpy as np
import pandas as pd
import datetime as dt
from dateutil.relativedelta import *
from pandas.tseries.offsets import *
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.iolib.summary2 import summary_col
import matplotlib.pyplot as plt
```

```python
wage=pd.read_csv('E:\Dr Ransom\wages.csv')
wage
```

```python
wage.describe()
```

```python
#Drop observations where either hgc or tenure are missing.
wage = wage.dropna(subset=['hgc', 'tenure'])
wage
```

```python
wage.describe()
```

```python
wage['Married_Ind'] = np.where(wage['married'] == 'married', 1, 0)
wage['College_Ind'] = np.where(wage['college'] == 'college grad', 1, 0)
wage
```

```python
wage.describe()
```

```python
# This one is a practice for the following imputation regression
#Use modelsummary to produce a summary table of this data frame.

model1 = sm.formula.ols('logwage ~ hgc + tenure + age + Married_Ind + College_Ind', data=wage).fit()
model2 = sm.formula.ols('logwage ~ hgc', data=wage).fit()
model3 = sm.formula.ols('logwage ~ tenure', data=wage).fit()
model4 = sm.formula.ols('logwage ~ age', data=wage).fit()
model5 = sm.formula.ols('logwage ~ Married_Ind', data=wage).fit()
model6 = sm.formula.ols('logwage ~ College_Ind', data=wage).fit()

# Create a list of models to display in the table
models = [model1, model2,model3, model4, model5, model6]

# Specify the column names and order
model_names = ['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5', 'Model 6']
column_order = ['coef', 'std err', 't', 'P>|t|', '95% Conf. Interval']

# Display the table
print(summary_col(models, model_names=model_names, float_format='%.2f', stars=True, regressor_order=column_order))
```

```python
#Question 7
wage['tenure_sq'] = wage['tenure'] ** 2
wage['college'] = (wage['hgc'] > 12).astype(int)
wage
```

```python

# Listwise deletion
complete_cases = wage.dropna()
model7 = sm.OLS(complete_cases['logwage'], 
                sm.add_constant(complete_cases[['hgc', 'college', 'tenure', 'tenure_sq', 'age', 'Married_Ind']])).fit()
print(model7.summary())

```

```python
# Mean imputation
mean_imputation = wage.fillna(wage.mean())
model8 = sm.OLS(mean_imputation['logwage'], 
                sm.add_constant(mean_imputation[['hgc', 'college', 'tenure', 'tenure_sq', 'age', 'Married_Ind']])).fit()
print(model8.summary())
```

```python
# Impute missing log wages as their predicted values from the complete cases regression
reg = LinearRegression().fit(complete_cases[['hgc', 'college', 'tenure', 'tenure_sq', 'age', 'Married_Ind']], 
                             complete_cases['logwage'])
predicted_values = reg.predict(wage[['hgc', 'college', 'tenure', 'tenure_sq', 'age', 'Married_Ind']])
predicted_imputation = wage.copy()
predicted_imputation['logwage'] = predicted_values
model9 = sm.OLS(predicted_imputation['logwage'], 
                sm.add_constant(predicted_imputation[['hgc', 'college', 'tenure', 'tenure_sq', 'age', 'Married_Ind']])).fit()
print(model9.summary())
```

```python
# Multiple imputation
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Set up the imputer
imp = IterativeImputer(max_iter=10, random_state=0)

# Impute the missing values
imputed = imp.fit_transform(wage[['logwage', 'hgc', 'college', 'tenure', 'tenure_sq', 'age', 'Married_Ind']])

# Convert the imputed array back to a DataFrame
wage_imputed = pd.DataFrame(imputed, columns=['logwage', 'hgc', 'college', 'tenure', 'tenure_sq', 'age', 'Married_Ind'])

# Fit the model
model10 = sm.OLS(wage_imputed['logwage'], 
                sm.add_constant(wage_imputed[['hgc', 'college', 'tenure', 'tenure_sq', 'age', 'Married_Ind']])).fit()

# Print the summary
print(model10.summary())
```

```python
models = [model7, model8,model9, model10]
# Specify the column names and order
model_names = ['Listwise deletion', 'Mean imputation', 'Missing Log', 'Multiple imputation']
column_order = ['coef', 'std err', 't', 'P>|t|', '95% Conf. Interval']

# Display the table
print(summary_col(models, model_names=model_names, float_format='%.3f', stars=True, regressor_order=column_order))
```
